```python
import json

essays = []
with open('../data/combined_essays.jsonl', 'r', encoding='utf-8') as f:
    for line in f:
        essays.append(json.loads(line))

print(f"Загружено {len(essays)} эссе")
```

```python
ivy_panda_essays = [essay for essay in essays if essay['source'] == 'IvyPanda Essays']
print(f"Загружено {len(ivy_panda_essays)} эссе с IvyPanda Essays\n")

print(f"Пример:\n{json.dumps(ivy_panda_essays[5], indent=2, ensure_ascii=False)}")
```

```python
import pandas as pd


rows = []
for essay in ivy_panda_essays:
    text = essay['text']
    title, _, body = text.partition('\n\n')

    if "Ukraine" in body:
        continue
    
    rows.append({"title": title, "text": body})

df = pd.DataFrame(rows)
df
```

```python
print(f"{df.iloc[5,0]}\n\n{df.iloc[5,1]}")
```

```python
df.to_csv("../datasets/human_essays.csv", index=False, encoding="utf-8")
```
